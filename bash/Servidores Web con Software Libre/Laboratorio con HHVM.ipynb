# Instalación de Php con Versión HHVM

Según estudiamos, Facebook creó su versión de Php fuera del proyecto original, con el fin de tener un mejor desempeño que Php versión 5 o 6.

Este servidor de HHVM (Hip Hop Virtual Machine) puede ser ejecutado sin necesidad de Apache, y en este tutorial lo veremos funcionando de ambas formas.


## Alistando el servidor

Comenzamos la instalación con el paquete de HHVM

```bash
sudo apt -yqq install hhvm
```

Comprobamos la versión instala de Php

```bash
php -v

# Definimos una variable, e imprimimos el contenido de la variable
php -r ' $hola_mundo = "Greencore\n"; print_r($hola_mundo); '

# Buscamos si existe un módulo de Php llamado "phpinfo"
php -r 'echo function_exists("phpinfo") ? "Sí\n" : "No\n";'
```

Seguimos por copiar contenido de ejemplo para probar el servidor

```bash
# Nótese que guardamos en archivo en nuestro directorio actual, nuestro HOME
echo "<?php phpinfo(); ?>" | tee index.php

# Con screen dejamos corriendo servidor sin que consuma la terminal
screen -S test -d -m sudo hhvm -m server -p 8080

# Probamos el acceso al contenido Php
curl http://localhost:8080/index.php
```

## Php HHVM con Apache2

Apache2 va a funcionar solo como un proxy hacia el puerto de HHVM (8080 en este ejemplo), usando el módulo de FastCGI para iniciar la conexión puerto de HHVM.

A futuro estudiaremon detalle el significado de la configuración del proxy.

```bash
# Matamos el servidor iniciado manualmente
sudo killall hhvm

# Instalamos Apache y habilitamos módulos
sudo apt-get install apache2 -yqq
sudo a2enmod proxy proxy_fcgi
echo 'ProxyPassMatch ^/(.+\.(hh|php)(/.*)?)$ fcgi://127.0.0.1:9000/var/www/html/$1' | sudo tee /etc/apache2/mods-enabled/hhvm_proxy_fcgi.conf
sudo systemctl restart apache2
sudo systemctl restart hhvm
```

Ahora comprobamos el acceso al contenido del sitio web, visitando http://IP-contenedor_HHVM/index.php o digitando:

```bash
curl http://localhost/index.php
```
