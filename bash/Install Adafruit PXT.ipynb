# Adafruit-PXT Local Dev Install

I'm trying to install a local version of Adafruit-PXT that can be used to teach in places without a permanent or good conection to the Internet.

The instructions are based on the [README.md](https://github.com/Microsoft/pxt-adafruit) of the Adafruit-PXT project.

This is being executed on a freshly installed Ubuntu 16.04 with updates and a ssh server.

### Upgrading

We start by updating the Debian repo and upgrading packages.

````bash
sudo apt-get update -yq
sudo apt-get dist-upgrade -y
````

### NodeJS

Next we install NodeJS version 8 from the nodesource script, as Ubuntu has an older version of NodeJS.

````bash
curl -sL https://deb.nodesource.com/setup_8.x | sudo -E bash -
sudo apt-get install -y nodejs
````

### Python build-dep packages

We need to install Yotta using Pip so first we install packages need to compile.

````bash
sudo apt-get install -y python-setuptools cmake build-essential ninja-build python-dev libffi-dev libssl-dev && sudo easy_install pip git
````

### Yotta

With Pip ready, we install yotta.

````bash
sudo pip install --upgrade yotta
````

### Github code

Now we download the code from Github

````bash
cd ~
git clone https://github.com/Microsoft/pxt.git
git clone https://github.com/Microsoft/pxt-common-packages.git
git clone https://github.com/Microsoft/pxt-adafruit.git
ls
````

### PXT Npm install

With the code ready, we proceed to "npm install" inside the PXT dir.

````bash
cd pxt
npm install -y
````

### Typings install

With NPM dependencies ready, we us the typing command to install

````bash
$HOME/pxt/node_modules/.bin/typings install
````

### Jake

Now we run Jake

````bash
$HOME/pxt/node_modules/.bin/jake
````

### PXT-common link

Now we do an NPM link command to the pxt-common-packages repo.

````bash
cd ../pxt-common-packages
npm install
sudo npm link ../pxt
````

### PXT-adafruit link

Now we link inside pxt-adafruit repo.

````bash
cd ../pxt-adafruit
npm install
sudo npm link ../pxt
sudo npm link ../pxt-common-packages
````

### Change path

It is important to change the PATH in the session to add the .bin directories, by running

```bash
export PATH=$PATH:$HOME/pxt/node_modules/.bin/:$HOME/pxt/pxt-cli
```

## Running PXT

Now, to run the server go to **a terminal** and write:

``cd ~/pxt-adafruit/ && pxt serve --cloud``
